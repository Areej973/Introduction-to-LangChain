### Prompt Templates

A prompt template refers to a reproducible way to generate a prompt. It contains a text string (“the template”), that can take in a set of parameters from the end user and generate a prompt.

The prompt template may contain:

- instructions to the language model,

- a set of few shot examples to help the language model generate a better response,

- a question to the language model.


```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI

# 1. Initialize the correct class for chat models
# All modern OpenAI models like gpt-4o-mini are chat models.
chat_model = ChatOpenAI(model="gpt-4.1-mini", temperature=0.9)

# 2. Define the prompt template
template = "Write a {adjective} poem about {subject}."
prompt_template = ChatPromptTemplate.from_template(template)
# No need to call .format, just for visualization
prompt_template.format(adjective='happy', subject='robots')

```

```python
# 3. Create the chain using the LCEL pipe operator
# This is the modern replacement for the old LLMChain.
chain = prompt_template | chat_model | StrOutputParser()

# 4. Invoke the chain with a dictionary of the input variables
response = chain.invoke({"adjective": "happy", "subject": "robots"})

print(response)
```

## Chains

Using an LLM in isolation is fine for some simple applications, but many more complex ones require chaining LLMs - either with each other or with other experts.

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

chat_model = ChatOpenAI(model="gpt-4.1-mini",temperature=1.0)

# --- First Chain: Name Generator ---
prompt1 = ChatPromptTemplate.from_template(
    "Brainstorm a name for a company that makes {product}."
    "The name should be evocative, modern, and suggest a connection to nature and technology. "
)
chain1 = prompt1 | chat_model | StrOutputParser()

# --- Second Chain: Description Generator ---
prompt2 = ChatPromptTemplate.from_template(
    "Write a one-sentence description for a company named '{company_name}'."
)
chain2 = prompt2 | chat_model | StrOutputParser()
```

```python
# first chain by itself ( several names How to fix?)

# Invoke the first chain to get a company name
company_name = chain1.invoke({"product": "eco-friendly water bottles"})

print("----------- OUTPUT OF CHAIN 1 -----------")
print(f"Generated Company Name: {company_name}")
```

```python
# Manually pass the output of chain1 as the input to chain2
description = chain2.invoke({"company_name": "AquaVera"})

print("----------- OUTPUT OF CHAIN 2 -----------")
print(f"Generated Description: {description}")
```

```python
# --- NEW: A function to parse the output of chain1 ---
def get_first_name(response: str) -> str:
  """Takes the raw LLM output and extracts the first company name."""
  for line in response.split('\n'):
    if line.strip().startswith(('1.','-','*')):
      return line.split('. ')[-1].strip()
  return "EcoPure"
```

```python
# --- Build the final chain with the parsing function included ---
sequential_chain = chain1 | get_first_name |(lambda name: {"company_name":name})|chain2
```

```python
# Invoke the full sequential chain
final_response = sequential_chain.invoke({"product": "water bottles"})
print("----------- OUTPUT OF THE COMBINED CHAIN -----------")
print(final_response)
```

## Agents & Tools

Agents use an LLM to determine which actions to take and in what order. An action can either be using a tool and observing its output, or returning to the user.

- Tool: A function that performs a specific duty. This can be things like: Google Search, Database lookup, Python REPL, other chains.

- LLM: The language model powering the agent.

- Agents: Agents involve an LLM making decisions about which Actions to take, taking that Action, seeing an Observation, and repeating that until done.

##### ***Potential use cases:***
- Personal Assistant
- Question Answering
- Chatbots
- Code Understanding etc.

### Memory


Memory is the concept of persisting state between calls of a chain/agent. LangChain provides a standard interface for memory, a collection of memory implementations, and examples of chains/agents that use memory.

```python
from langchain_openai import ChatOpenAI
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
llm = ChatOpenAI(model="gpt-4.1-mini", temperature=0)
store = {}
def get_session_history(session_id: str) -> ChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a friendly and talkative AI assistant."),
    MessagesPlaceholder(variable_name="history"),
    ("user", "{input}")
])
chain = prompt | llm
conversation_with_memory = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="history",
)

print("Conversational chain with memory is ready.")
```

```python
session_id = "my_first_conversation"

print("--- Turn 1 ---")
response1 = conversation_with_memory.invoke(
    {"input": "Hi! My name is Alex. Tell me a joke"},
    config={"configurable": {"session_id": session_id}}
)
print(f"AI: {response1.content}")

print("\n--- Turn 2 ---")
response2 = conversation_with_memory.invoke(
    {"input": "What's my name?"},
    config={"configurable": {"session_id": session_id}}
)
print(f"AI: {response2.content}")

print("\n--- Turn 3 ---")
response3 = conversation_with_memory.invoke(
    {"input": "Tell me a similar joke"},
    config={"configurable": {"session_id": session_id}}
)
print(f"AI: {response3.content}")

print("\n--- Current Memory for session '{}' ---".format(session_id))
print(store[session_id].messages)
```

### Indexes

Indexes refer to ways to structure documents so that LLMs can best interact with them. This module contains utility functions for working with documents, different types of indexes, and then examples for using those indexes in chains.

- Embeddings: An embedding is a numerical representation of a piece of information, for example, text, documents, images, audio, etc.
- Text Splitters: When you want to deal with long pieces of text, it is necessary to split up that text into chunks.
- Vectorstores: Vector databases store and index vector embeddings from NLP models to understand the meaning and context of strings of text, sentences, and whole documents for more accurate and relevant search results.


```python
import requests

url = "https://s3.amazonaws.com/weclouddata/datasets/genai/langchain/state_of_the_union.txt"
res = requests.get(url)
with open("state_of_the_union.txt", "w") as f:
  f.write(res.text)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('./state_of_the_union.txt')
documents = loader.load()
```

```python
# Text Splitter
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
len(docs)
```

```python
# Embeddings

from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import FAISS  # moved in v1.x

db = FAISS.from_documents(docs, embeddings)

query = "What did the president say about the Supreme Court"
docs = db.similarity_search(query)
```
